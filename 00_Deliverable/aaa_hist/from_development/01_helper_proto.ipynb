```python
import cv2
import os
import numpy as np


def data_augmentation(input_folder, output_folder, target_count):
    # Create the output folder if it doesn't exist
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    # Get a list of all image file names in the input folder
    image_files = [f for f in os.listdir(input_folder) if f.endswith('.jpg') or f.endswith('.png')]

    # Calculate the number of images to generate from each input image
    augmentation_factor = target_count // len(image_files)

    # Iterate over each image file
    for image_file in image_files:
        # Load the input image
        image_path = os.path.join(input_folder, image_file)
        img = cv2.imread(image_path)

        # Apply data augmentation techniques
        augmented_images = []

        # Perform rotation
        for angle in range(120, 359, 120):
            rotation_matrix = cv2.getRotationMatrix2D((img.shape[1] / 2, img.shape[0] / 2), angle, 1)
            rotated_img = cv2.warpAffine(img, rotation_matrix, (img.shape[1], img.shape[0]))
            augmented_images.append(rotated_img)

        # Perform horizontal and vertical flips
        flipped_img = cv2.flip(img, 0)
        augmented_images.append(flipped_img)
        flipped_img = cv2.flip(img, 1)
        augmented_images.append(flipped_img)

        # Perform shifting
        num_pixels = 10
        translation_matrix = np.float32([[1, 0, num_pixels], [0, 1, num_pixels]])
        translated_img = cv2.warpAffine(img, translation_matrix, (img.shape[1], img.shape[0]))
        augmented_images.append(translated_img)

        # Save the augmented images to the output folder
        base_filename = os.path.splitext(image_file)[0]
        for i, augmented_img in enumerate(augmented_images):
            output_filename = f"{base_filename}_{i}.jpg"
            output_path = os.path.join(output_folder, output_filename)
            cv2.imwrite(output_path, augmented_img)

    print(f"Data augmentation complete! Generated {target_count} images.")


# Usage example
input_folder = '/home/dkhatanassia/detectron2_dataset/jerry_real/imgs'
output_folder = '/home/dkhatanassia/detectron2_dataset/jerry_real/imgs_aug'
target_count = 2400
data_augmentation(input_folder, output_folder, target_count)
renameImgs(output_folder,375)
lala = initDict(output_folder,375)

with open(json_path_mc_tr, 'r') as f:
    base= json.load(f)

base['images']+= lala['images']
    # save new file
with open(json_path_mc_tr, 'w') as f:
    json.dump(base, f)
```

```python
import os
from PIL import Image

def resize_and_copy_images(input_folder, output_folder, H, W):
    # Create the output folder if it doesn't exist
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    # Get the list of image files in the input folder
    image_files = [f for f in os.listdir(input_folder) if f.endswith('.jpg') or f.endswith('.png')]

    # Process each image
    for file in image_files:
        # Open the image file
        image_path = os.path.join(input_folder, file)
        image = Image.open(image_path)

        # Resize the image
        resized_image = image.resize((W, H))

        # Save the resized image to the output folder
        output_path = os.path.join(output_folder, file)
        resized_image.save(output_path)

        print(f"Image '{file}' resized and copied successfully.")

# Example usage
input_folder = '/home/dkhatanassia/detectron2_dataset/jerry_real/_imgs'
output_folder = '/home/dkhatanassia/detectron2_dataset/jerry_real/imgs'
H = 720  # Desired height
W = 1280  # Desired width

resize_and_copy_images(input_folder, output_folder, H, W)
```
