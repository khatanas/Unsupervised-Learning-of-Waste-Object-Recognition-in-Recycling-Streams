```python
def correctFormat2(path_file_video,path_root = path_root_videos):
    tail = path_file_video.replace(path_root+'/','')
    print(tail)
    splitted = tail.split('/')
    channel = splitted[0]
    name = splitted[-1]
    print(splitted)
    if len(splitted) != 5:
        return False
    
    else:
        #CH00001/2022/10/21/Argos_00_20221021075511.mp4
        #IN00001/2023/03/23/Argos_00_20230323194402.mp4
        if channel in ['CH00001','IN00001']:
            splitted = name.split('_')
            if len(splitted) != 3: return False
            else: return True if all([len(item)==ref for item, ref in zip(splitted,[6,2,18])]) else False
            
        #CH00004/2023/03/02/2023-03-02T07%3A35%3A34.017065.avi'
        elif channel == 'CH00004':
            return True if len(name)=34 else False
        
        else: return False
```

```python
# empty
visualizeSA(image,title=f'{id_cross}.jpg\n',path_save=path_dir_save)

# coarse points 
title=f'{id_cross}.jpg\n Coarse SA automatic prompt points'
prompts = np.array([i['point_coords'][0] for i in masks_coarse])
plt.figure(figsize=(p,p))
plt.title(title)
plt.imshow(image)
show_points(prompts,np.ones(len(prompts)),plt.gca())
plt.axis('off')
if path_dir_save != False:plt.savefig(join(path_dir_save,f'{title}.jpg'),bbox_inches='tight')
plt.show()

# coarse masks
visualizeSA(image, [mask for mask in masks_coarse], title=f'{id_cross}.jpg\n Coarse masks')

# fine points
title=f'{id_cross}.jpg\n Fine SA automatic prompt points'
prompts = np.array([i['point_coords'][0] for i in masks_fine])
plt.figure(figsize=(p,p))
plt.title(title)
plt.imshow(image)
show_points(prompts,np.ones(len(prompts)),plt.gca())
plt.axis('off')
if path_dir_save != False:plt.savefig(join(path_dir_save,f'{title}.jpg'),bbox_inches='tight')
plt.show()

# not crossing fine masks
visualizeSA(image, [mask for mask in masks_fine if mask['AoI']==1],title=f'{id_cross}.jpg\n Fine masks')

# final masks 
visualizeSA(image, [mask for mask in masks_fine],title=f'{id_cross}.jpg\n Final masks')

# final objects
visualizeOnly(image, [mask for mask in masks_fine],title=f'{id_cross}.jpg\n Final objects',value=255)
```

```python
k = 3

for cat in ['plastic drum']:#category_list:
    path_dir_save = join(path_root_save,'classification',cat)
    os.makedirs(path_dir_save,exist_ok=True)
    #path_dir_save=False
    
    print(cat)
    # path to directory where preview images for specific cat are stored
    path_dir_selected_images = join(path_root_images_selection,sampling_pts,channel,str(threshold),f'images_preview_{cat}_{threshold}')
    # path_list of all images in dir
    tmp_list_path_images = collectPaths(path_dir_selected_images)
    # subset of 5
    path_list_k = random.sample(tmp_list_path_images,min(k, len(collectPaths(tmp_list_path_images))))
    # image_id of subset
    ids_k = [getId(p.split('/')[-1][4:]) for p in path_list_k]
    
    # for each image in subset
    for p,id in zip(path_list_k,ids_k):
        print(id)
        # get masks file
        path_file_mask_tmp = getMasksPath(id,sampling_pts)
        
        # remove those which weren't use for classification
        mask_ids = [ann['id'] for ann in classification_dico 
                    if ann['image_id'] == id and 
                    ann['name'] == cat and 
                    ann['score'] >= threshold]
        masks = decodeMasks(readJson(getMasksPath(id,sampling_pts)))
        classification_masks = [mask for mask in masks if mask['id'] in mask_ids]
        
        image = imreadRGB(p)
        visualizeSA(image, title=f'{p.split("/")[-1][4:]}\n',path_save=path_dir_save)
        visualizeSA(image,classification_masks, title = f'{p.split("/")[-1][4:]}\n CLIP score "{cat}" >=  {threshold}',path_save=path_dir_save)
```
